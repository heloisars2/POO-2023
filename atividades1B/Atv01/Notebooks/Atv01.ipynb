Apresente a soma de todos os argumentos inteiros ou reais recebidos

```java
import java.util.Scanner;

Scanner h = new Scanner(System.in);
        
System.out.println("Digite dois valores para serem somados:");
int num_1 = h.nextInt(); 
int num_2 = h.nextInt(); 
        
h.close();
        
int soma = num_1 + num_2;
        
System.out.println("A soma dos argumentos é : " + soma);
```
